```python
import pandas as pd
import datetime
```

```python
data_o = r'https://github.com/06Cata/Analyzing_wind_speed_for_Ireland/blob/main/ireland_wind.txt'
data = pd.read_csv(data_o, sep = '\s+', parse_dates = [[0, 1, 2]])
data.head(3)

# data
# 出來結果是 2070 ~ 2072、1973 ~ 1978
```

```python
# 調整跑掉的日期
def fix(x):
    if x.year >= 2070:
        year = x.year - 100
    else:
        year = x.year
    return datetime.datetime(year, x.month, x.day)

data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix)

data

# data.to_csv('output.csv', index=False)
```

```python
# apply 練習
# data.apply(代入def名稱)
# 前面x是一個參數, 輸入每一個data值
import pandas as pd

data_apply = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

def multiply_by_two(x):
    return x * 2

result = data_apply.apply(multiply_by_two)

print(result)
```

```python
import matplotlib.pyplot as plt

plt.figure(1)

# 增加年份欄位，計算年份平均值
data['Year'] = data['Yr_Mo_Dy'].dt.year
rpt_mean_by_year = data.groupby('Year')['RPT'].mean() 

rpt_mean_by_year.plot(kind='line', marker='o') 
plt.xlabel('Year')
plt.ylabel('RPT Mean') 
plt.title('Average RPT by Year')
plt.show()  

# PS 數據點可以改 
# 's'：正方形
# 'D'：菱形
# '^'：上三角形
# 'v'：下三角形
# '+'：加号
# '*'：星号
# 'x'：叉号
```

```python
plt.figure(2)

# 提取列名
cities = data.columns[1:]

# 六月平均
avg_june = data.loc[data.index.month == 6].mean()

plt.plot(cities, avg_june[1:], marker='o')

plt.title('Average Values in June')
plt.xlabel('City')
plt.ylabel('Average Value')

# x座標的字，旋轉45度
plt.xticks(rotation=45)


plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(3)

mean_4 = data.loc[data.index.month == 4].mean()
mean_5 = data.loc[data.index.month == 5].mean()
mean_6 = data.loc[data.index.month == 6].mean()

combined_mean= pd.concat([mean_4, mean_5, mean_6], axis=1)
combined_mean.columns = ['April', 'May', 'June']  # 設置列名

numeric_combined_mean = combined_mean.drop(['Yr_Mo_Dy','Year', 'Week', 'Weekday'], axis=0)

numeric_combined_mean.plot(kind='line', marker='o')

plt.title('Average Values from April to June')
plt.xlabel('City')
plt.ylabel('Average Value')

plt.xticks(rotation=45)

plt.show()
```

```python
# 以每年每個月為群組的平均
# data.index.to_period(' ') 以某時間周期為單位

data.groupby(data.index.to_period('M')).mean()
```

```python
# 以每年每個月為群組的詳細統計資料
data.groupby(data.index.to_period('M')).describe()
```

```python
# 以每年每個月為群組的mean, min, max,.agg() 聚合
ym = data.groupby(data.index.to_period('M')).agg(['mean', 'min', 'max'])
ym
```

```python
# 以每年的同一個月為群組的mean, min, max
data.groupby(data.index.month).agg(['mean', 'min', 'max'])
```

```python
# 以每年每週為群組的詳細統計資料
data.groupby(data.index.to_period('W')).mean()
```

```python
# 以每年的同一週為群組的mean, min, max
# data.index.isocalendar()
data.groupby(data.index.isocalendar().week).agg(['mean', 'min', 'max'])
```

```python
iso_calendar = data.index.isocalendar()  

data['Year'] = iso_calendar['year']
data['Week'] = iso_calendar['week']
data['Weekday'] = data.index.dayofweek

result = data.groupby(['Year', 'Week', 'Weekday']).agg(['mean', 'min', 'max'])

print(result)
```
